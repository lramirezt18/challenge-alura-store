### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_t1 = tienda["Precio"].sum()
ingreso_t2 = tienda2["Precio"].sum()
ingreso_t3 = tienda3["Precio"].sum()
ingreso_t4 = tienda4["Precio"].sum()
```

```python
print("Ingreso total Tienda 1:", ingreso_t1)
print("Ingreso total Tienda 2:", ingreso_t2)
print("Ingreso total Tienda 3:", ingreso_t3)
print("Ingreso total Tienda 4:", ingreso_t4)
```

```python
import matplotlib.pyplot as plt

ingresos = {
    "Tienda 1": tienda["Precio"].sum(),
    "Tienda 2": tienda2["Precio"].sum(),
    "Tienda 3": tienda3["Precio"].sum(),
    "Tienda 4": tienda4["Precio"].sum()
}

plt.figure()
plt.bar(ingresos.keys(), ingresos.values(),
        color=["blue","orange","red","yellow"])

plt.title("Ingresos Totales por Tienda")
plt.xlabel("Tiendas")
plt.ylabel("Ingresos")

plt.show()
```

# 3. Calificación promedio de la tienda


### Calcular promedio de calificación por tienda

```python
promedio_t1 = tienda["Calificación"].mean()
promedio_t2 = tienda2["Calificación"].mean()
promedio_t3 = tienda3["Calificación"].mean()
promedio_t4 = tienda4["Calificación"].mean()
```

### Mostrar resultados

```python
print("Valoración promedio Tienda 1:", round(promedio_t1, 2))
print("Valoración promedio Tienda 2:", round(promedio_t2, 2))
print("Valoración promedio Tienda 3:", round(promedio_t3, 2))
print("Valoración promedio Tienda 4:", round(promedio_t4, 2))
```

```python
calificaciones = {
    "Tienda 1": tienda["Calificación"].mean(),
    "Tienda 2": tienda2["Calificación"].mean(),
    "Tienda 3": tienda3["Calificación"].mean(),
    "Tienda 4": tienda4["Calificación"].mean()
}

plt.figure()

plt.plot(
    list(calificaciones.keys()),
    list(calificaciones.values()),
    marker="o",
    color="purple",
    linewidth=3
)

plt.title("Calificación Promedio por Tienda")
plt.xlabel("Tiendas")
plt.ylabel("Calificación")

plt.show()
```

# 5. Envío promedio por tienda

####  En esta sección analizamos cuánto pagan los clientes en promedio por el envío de sus compras en cada tienda. Utilizamos un diccionario para organizar los resultados y facilitar la comparación entre las diferentes tiendas.


```python
# Crear un diccionario con las tiendas y sus datos
tiendas = {
    "Tienda 1": tienda,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}

# Calcular el costo de envío promedio por tienda
envio_promedio = {}

for nombre, df in tiendas.items():
    envio_promedio[nombre] = df["Costo de envío"].mean()

# Mostrar los resultados
for tienda_nombre, promedio in envio_promedio.items():
    print(f"Costo de envío promedio - {tienda_nombre}: {promedio:.2f}")
```
